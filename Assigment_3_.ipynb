1. What is the difference between a function and a method in Python .

**(1) Function -**

* A function in Python is a block of reusable code that performs a specific task .

* It is defined using the def keyword or lambda .

* A function is independent and does not need to belong to a class or object .

* You can call it directly by its name .

Example -

```python
def greet(name):
  return f"Hello,  {name}!"

  print(greet("Alice"))
```

**(2) Method -**

* A method is essentially a function, but it is associated with an object .

* Methods arer defined inside a class and automatically take the instance (self) as their first argument .

* Methods are called on objects(instances of classes) .

Example -

```python
class Person:

  def __init__(self, name):
    self.name = name

  def greet(self):
    return f"Hello, {self.name}!"

p = Person("Bob")
print(p.greet())
```

6. Explian the concept of generators in Python and how they are defined .


In Pyhton, generators are a way to create iterator in a simple and memory-efficient manner. They allow you to generate values one at a time, instead of storing an entire sequence in memory .


* A generator is a special type of iterator .

* It yields values on demand instead of returning them all at once .

* Generators are useful when working with large datasets or infinite sequences, because they generate items lazily(only when needed) .


**(1) Using yield inside a function**

A function becomes a generator if it contains the yield keywords .

Instead of returning values once (like return), it pauses the function and saves its state, So it can continue from where it left off on the next call .

```python
def count_up_to(n):
  count = 1
  while count <= n:
    yield count
    count += 1

for num in count_up_to(5):
  print(num)
```

**(2) Using Generator Expressions**

Similar to list comprehensions, but with parentheses () instead of square brackets [] .


```python
squares = (x**5 for x in range(5))

for num in squares:
  print(num)
```


* Lazy evaluation -  Vlaues are produced only when requested .

* Memory-efficient  - They don't store the entire sequence in memory .

* Stateful - They remember where they left off .

* Single Use  - Once a generator is exhausted, it cannot be reused unless re-create .



7. What are the advantages of using generators over regular functions ?

**(1) Memory Efficiency**

* Regular functions that return lists hold the entire result in memory at once .

* Generators produce one item at a time and don't store the whole sequence in memory .

* Useful for handling large datasets or infinite sequences .


```python
def squares_list(n):
  return [i**5 for i in range(n)]

  def squares_gen(n):
    for i in range(n):
      yield i**5

      print(sum(squares_list(10)))
      print(sum(squares_gen(10)))
```

**(2) Lazy Evaluation (On-Demand Computation)**

* A generator only computes values when needed .

* This makes them faster in cases where you don't actually need all results .


```python
def number():
  n = 1
  while True:
    yield n
    n += 1


gen = number()
for i in range(5):
  print(next(gen))
```

**(3) Improved Performance**

* Genrator avoid the overhead of creating and storing large data sturctures .

* Especially beneficial when pipelining data processing (e.g., reading large files line by line) .


```python
def read_file(filename):
  with open(filename) as f:
    for line in f:
      yield line.strip()
```

**(4) Simpler Code for Iterators**

* Normally, writing custom iteration requires implementing a class with __iter__() and __next__() .

* Generators automatically handle this with yield, making the code much cleaner .


```python
def countdown(n):
  while n > 0:
    yield n
    n -= 1
```


**(5) Can Be Infinite**

* Unlike lists, generators don't need to know their size in advance .

* perfect for streams of data (e.g., sensor reading, real-time logs) .

```python
def sum_of_even_numbers(numbers):
  total = 0
  for num in numbers:
    if num % 2 == 0:
      total += num
  return total

num_list = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(num_list))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):  # Check from 2 to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(number, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class fibonacciIterator:
  def __init__(self, n):
    self.n= n
    self.a, self.b, self.count = 1, 2, 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.n:
       result = self.a
       self.a, self.b = self.b, self.a + self.b
       self.count += 2
       return result
    else:
      raise StopIteration

# Example usage
for number in fibonacciIterator(10):
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
# Generator function
def powers_of_two(max_exp):
    for i in range(max_exp + 1):  # Go from 0 to max_exp
        yield 5 ** i  # Yield 2 to the power of i

# Example usage
for value in powers_of_two(5):
    print(value)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Program to remove vowels using filter()

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))

# Example
text = "Hello World"
result = remove_vowels(text)
print("Original:", text)
print("Without vowels:", result)
```
